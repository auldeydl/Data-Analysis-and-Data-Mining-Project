## 各行业发生股利分配公司占比

```python
#公司去重统计行业分布
def qvchongtongji(data):
    company = data.drop_duplicates(['证券简称'])
    coindustries_series = company['一级行业名称'].value_counts()
    return coindustries_series

def gulizhanbi(a,b):
    c = qvchongtongji(a)
    d = qvchongtongji(b)
    df = pd.concat([c,d, d/c],join='outer', axis=1)
    df.columns = ['上市公司行业分布数目','有股利分配的上市公司行业数目','占比']
    df = df.sort_values(by = '占比',ascending=False)
    return df
```

# 各行业股利分配方式（转增、送股、派息、派息并转增、派息并送股等）分析

```python
def fenpeifangshi(data):
    data['股利分配方式'] = ''
    da = data[['证券简称','转增比','派息比税前','送股比','股利分配方式']]
    da = da.fillna(-1)
    n = da.shape[0]
    da.reset_index(drop=True, inplace=True)
    for i in range(n):
        a = da.loc[i]['派息比税前']
        b = da.loc[i]['送股比']
        c = da.loc[i]['转增比']
        if a>0 and b>0 and c>0:
            da.loc[i,'股利分配方式'] = '派息、转增并送股'
        elif a>0 and b>0 and c<0:
            da.loc[i,'股利分配方式'] = '派息并送股'
        elif a>0 and b<0 and c>0:
            da.loc[i,'股利分配方式'] = '派息并转增'
        elif a<0 and b>0 and c>0:
            da.loc[i,'股利分配方式'] = '送股并转增'
        elif a>0 and b<0 and c<0:
            da.loc[i,'股利分配方式'] = '派息'
        elif a<0 and b<0 and c>0:
            da.loc[i,'股利分配方式'] = '转增'
        else:
            da.loc[i,'股利分配方式'] = '送股'
    df = pd.DataFrame(da.groupby('股利分配方式').size(),columns=[data.iloc[520]['财政年度']])
    return df
```

```python
df = pd.concat([fenpeifangshi(data10),fenpeifangshi(data11),fenpeifangshi(data12),
                fenpeifangshi(data13),fenpeifangshi(data14),fenpeifangshi(data15),
                fenpeifangshi(data16),fenpeifangshi(data17),fenpeifangshi(data18)]
               ,axis = 1)
df
```

# 股利连续分配情况（1年、2年、3年、4年、5年）

```python
def continusFind(num_list):
    s=1
    find_list=[]
    have_list=[]
    while s<=len(num_list)-1:
        if num_list[s]-num_list[s-1]==1:
            flag=s-1
            while (s <= (len(num_list)-1)) and (num_list[s]-num_list[s-1]==1):
                s+=1
            find_list.append(num_list[flag:s])
            have_list+=num_list[flag:s]
        else:
            s+=1
    return find_list

def fenpei(d):
    dataa = d[['证券简称','财政年度']]
    c = dataa.drop_duplicates(['证券简称','财政年度'])
    c.index  = c['财政年度']
    groups = c.groupby('证券简称').groups
    lianxvguli = dict.fromkeys(list(groups.keys()),0)

    for i in range(len(groups)):
        num_list = list(groups[list(groups.keys())[i]].values)
        find_list = continusFind(num_list)
        if len(find_list)==0:
            lianxvguli[list(c.groupby('证券简称').groups.keys())[i]] =1
        else:
            lianxvguli[list(c.groupby('证券简称').groups.keys())[i]]= len(find_list[0])
    return lianxvguli
```

```python
df = pd.DataFrame(pd.Series(fenpei(data2)))
df = df.reset_index()
df.columns= ['证券简称','连续分配年数']

df2 = data2[['证券简称','一级行业名称']]
df2 = df2.drop_duplicates(['证券简称'])

df3 = df.merge(df2,how = 'left')
```
